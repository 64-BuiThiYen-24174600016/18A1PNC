```python
class Date:
    def __init__(self, day:int, month:int, year:int):
        self.day = day
        self.month = month
        self.year = year

    def display(self):
        print(f"{self.day:02d}/{self.month:02d}/{self.year}")
```

```python
d1 = Date(6, 10, 2025)
d1.display()

#thiáº¿u next
```
