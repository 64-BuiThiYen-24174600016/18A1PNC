```python
class PS:
    def __init__(self, tu:int, mau:int):
        self.tu = tu
        self.mau = mau

    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))

    def display(self):
        print(f"Phân số: {self.tu}/{self.mau}")
```

```python
ps1 = PS(1, 2)
ps1.display()
```
