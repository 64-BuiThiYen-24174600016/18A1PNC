```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[-1]

    #trả về số phần tử trong stack
    def count(self):
        return len(self.items)
    
```

```python
#Số phần tử
s = Stack()
s.push(10)
s.push(20)
s.push(30)
s.count()
```

```python
#Số phần tử sau pop
s.pop()
s.count()
```
