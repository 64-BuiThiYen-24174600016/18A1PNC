```python
class Stack:
    def __init__(self):
        self.items = []  # danh sách phần tử trong ngăn xếp

    def push(self, item):
        #Thêm phần tử vào đỉnh ngăn xếp
        self.items.append(item)

    def pop(self):
        #Lấy phần tử trên cùng của ngăn xếp
        return self.items.pop()

    def peek(self):
        #Xem phần tử trên cùng mà không lấy ra
        return self.items[-1]

    def print(self):
        print("Nội dung ngăn xếp:", self.items)
```

```python
s = Stack()
s.push("A")
s.push("B")
s.push("C")
s.print() 
```

```python
#sau pop
s.pop()
s.print()
```
