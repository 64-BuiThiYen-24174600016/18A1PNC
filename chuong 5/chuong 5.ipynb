# Bài 5.10


```python
import sqlite3

def connect_db():
    return sqlite3.connect("product.db")

def create_table():
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS product (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT NOT NULL,
            price REAL NOT NULL,
            amount INTEGER NOT NULL
        )
    """)
    conn.commit()
    conn.close()

def show_products():
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM product")
    rows = cursor.fetchall()
    print("\nDanh sách sản phẩm:")
    for row in rows:
        print(row)
    conn.close()

def add_product(name, price, amount):
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute(
        "INSERT INTO product (name, price, amount) VALUES (?, ?, ?)",
        (name, price, amount)
    )
    conn.commit()
    conn.close()

def find_product(name):
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute(
        "SELECT * FROM product WHERE name LIKE ?",
        ('%' + name + '%',)
    )
    rows = cursor.fetchall()
    for row in rows:
        print(row)
    conn.close()

def update_product(product_id, price, amount):
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("""
        UPDATE product
        SET price = ?, amount = ?
        WHERE id = ?
    """, (price, amount, product_id))
    conn.commit()
    conn.close()

def delete_product(product_id):
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("DELETE FROM product WHERE id = ?", (product_id,))
    conn.commit()
    conn.close()
```

```python

create_table()

add_product("Laptop", 15000000, 10)
add_product("Kiên", 300000, 50)

show_products()

print("\nTìm sản phẩm chứa 'Lap':")
find_product("Lap")

update_product(1, 14000000, 8)
delete_product(2)

show_products()
```

# Bài 5.11

```python
import sqlite3

conn = sqlite3.connect("nhan_vien.db")
cursor = conn.cursor()

cursor.execute("""
    CREATE TABLE IF NOT EXISTS PHONG (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT NOT NULL
    )
""")

cursor.execute("""
    CREATE TABLE IF NOT EXISTS NHAN_VIEN (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        ho_ten TEXT NOT NULL,
        tuoi INTEGER NOT NULL,
        dia_chi TEXT NOT NULL,
        luong REAL NOT NULL,
        id_phong INTEGER,
        FOREIGN KEY (id_phong) REFERENCES PHONG(id)
    )
""")

conn.commit()

cursor.execute("INSERT INTO PHONG (name) VALUES (?)", ("Bảo vệ",))
cursor.execute("INSERT INTO PHONG (name) VALUES (?)", ("Vệ sinh môi trường",))

cursor.execute("""
    INSERT INTO NHAN_VIEN (ho_ten, tuoi, dia_chi, luong, id_phong)
    VALUES (?, ?, ?, ?, ?)
""", ("Trần Trung Kiên", 18, "Nam Định", 15000000, 1))

cursor.execute("""
    INSERT INTO NHAN_VIEN (ho_ten, tuoi, dia_chi, luong, id_phong)
    VALUES (?, ?, ?, ?, ?)
""", ("Nguyễn Đức Minh", 35, "Ninh Bình", 18000000, 2))

conn.commit()

cursor.execute("""
    SELECT NV.ho_ten, NV.tuoi, NV.luong, P.name
    FROM NHAN_VIEN NV
    JOIN PHONG P ON NV.id_phong = P.id
""")

print("\nDanh sách nhân viên:")
for row in cursor.fetchall():
    print(row)

cursor.close()
conn.close()
```
